# **Rock vs Mine Sonar Prediction**

## Importing necessary modules

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

## Loading data into a DataFrame

```python
df = pd.read_csv('/content/Copy of sonar data.csv',header = None)
```

## Exploring the data

```python
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df[60].value_counts()
```

```python
df.groupby(60).mean()
```

## Splitting the target variable

(Whether the object is Rock 'R' or Mine 'M', in this case)

```python
x = df.drop(columns = 60)
y = df[60]
```

```python
print(x,y)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size = 0.1, random_state = 2)
```

## Spitting x and y into training and testing data

```python
print(x_train,x_test,y_train,y_test)
```

## Model fitting

```python
logistic_model = LogisticRegression()
```

```python
logistic_model.fit(x_train, y_train)
```

## Checking the accuracy of the model

```python
x_train_prediction = logistic_model.predict(x_train)
acc_score_train = accuracy_score(x_train_prediction, y_train)
```

```python
print(acc_score_train)
```

```python
x_test_prediction = logistic_model.predict(x_test)
acc_score_test = accuracy_score(x_test_prediction, y_test)
print(acc_score_test)
```

# **Diabetes Prediction**

## Importing necessary modules

```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
```

## Loading the data into a DataFrame

```python
df = pd.read_csv('/content/diabetes.csv')
```

## Exploring the data

```python
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df['Outcome'].value_counts()
```

```python
df.groupby(df['Outcome']).mean()
```

## Splitting the target variable

```python
x = df.drop(columns = 'Outcome', axis = 1)
y = df['Outcome']
```

```python
print(x,y)
```

## Data Preprocessing

### Data Standardization

```python
scaler = StandardScaler()
```

```python
standardized_data = scaler.fit_transform(x)
```

```python
x = standardized_data
```

Splitting data into training and testing data

```python
x_train, x_test, y_train, y_test = train_test_split(x , y , test_size = 0.2 ,stratify = y, random_state = 2)
```

```python
print(x_train,x_test,y_train,y_test)
```

## Training the model

```python
classifier = svm.SVC(kernel = 'linear')
```

```python
classifier.fit(x_train, y_train)
```

## Model Evaluation

### Accuracy score

```python
x_train_prediction = classifier.predict(x_train)
acc_x_train = accuracy_score(x_train_prediction , y_train)
print(acc_x_train)
```

```python
x_test_prediction = classifier.predict(x_test)
acc_x_test = accuracy_score(x_test_prediction , y_test)
print(acc_x_test)
```

### Result verification

```python
input_data = (2,197,70,45,543,30.5,0.158,53)
array_input = np.asarray(input_data)
reshaped = array_input.reshape(1,-1)
std_data = scaler.transform(reshaped)
prediction1 = classifier.predict(std_data)
print(prediction1)
```

```python
input_data = (4,110,92,0,0,37.6,0.191,30)
array_input = np.asarray(input_data)
reshaped = array_input.reshape(1,-1)
std_data = scaler.transform(reshaped)
prediction2 = classifier.predict(std_data)
print(prediction2)
```
